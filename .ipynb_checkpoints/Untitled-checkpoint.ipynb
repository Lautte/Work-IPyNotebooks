```python
pylab.plot(EC, TDS, 'o')
#pylab.xlim(100, 400)
#pylab.ylim(100,350)
```

```python
# calc the trendline (it is simply a linear fitting)
z = numpy.polyfit(EC, TDS, 1)
p = numpy.poly1d(z)
pylab.plot(EC,p(EC),'r-')
# the line equation:
print 'y=%.6fx+(%.6f)'%(z[0],z[1])
```
